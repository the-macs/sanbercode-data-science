```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
data = ['tan', 'sin', 'cos']
colors = ['b', 'r', 'g']

fig, ax = plt.subplots(figsize=(20,10), nrows=1, ncols=3, sharex=True, sharey=True)
for i in range(len(data)):
    ax[i].plot(x, eval(data[i]), color=colors[i], label=f"{data[i]}X")
    ax[i].title.set_text(f"Grafik {data[i]}x")
    
fig.legend(loc="upper center", bbox_to_anchor=(0.5, 0., 0, 1.1))

plt.tight_layout()
plt.show();
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(20,10))
# x = np.linspace(0, 10, 20)
ax.plot(x, y)
ax.title.set_text("Grafik CosX")

ins1 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins1.plot(x, y5, color="y")
ins1.plot(x, y, color="b")
ins1.title.set_text("Grafik CosX^5")

ins2 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins2.plot(x, y4, color="k")
ins2.plot(x, y, color="b")
ins2.title.set_text("Grafik CosX^4")

ins3 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins3.plot(x, y3, color="g")
ins3.plot(x, y, color="b")
ins3.title.set_text("Grafik CosX^3")


ins4 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins4.plot(x, y2, color="r")
ins4.plot(x, y, color="b")
ins4.title.set_text("Grafik CosX^2")

# ins2 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
# ins2.plot(x, np.tan(x))

plt.show()
```
